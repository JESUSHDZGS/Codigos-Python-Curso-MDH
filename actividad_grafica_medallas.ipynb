```python
df=pd.read_csv('juegos_olimpicos.csv')
```

```python
df.head(2)
```

```python
oro=df['Medalla']=='Gold'
plata=df['Medalla']=='Silver'
bronce=df['Medalla']=='Bronze'
```

```python
medallero=df[oro|plata|bronce]
```

```python
federaciones=medallero['NOC'].value_counts()
```

```python
top_diez=federaciones[0:10]
top_diez.sort_values()
```

```python
sns.barplot(x=top_diez.index,y=top_diez.values,data=medallero)
```
