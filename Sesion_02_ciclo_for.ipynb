```python
# un programa donde le piden al usuario el limite superior
# para realizar la suma partiendo desde cero y hasta llegar
# al limite superior
# si el usuario me dice que el limite superior es 145
# suma de 1+2+3+4+...+145
```

```python
print('Realicemos una suma progresiva de numeros partiendo de cero')
limite=int(input('Elige el numero hasta el que quieres sumar '))
suma=0
for i in range(1,limite+1):
  suma+=i
print(suma)
```
