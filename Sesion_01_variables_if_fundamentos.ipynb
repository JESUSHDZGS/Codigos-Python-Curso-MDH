```python
print('Esta es una calculadora')
print('Elije el tipo de oepracion del  menu')
print('1. Sumar')
print('2. Restar')
print('3. Multiplicar')
print('4. Dividir')
resp=input('¿Que quieres hacer? ')
num1=float(input('Escribe el primer numero: '))
num2=float(input('Escribe el seugndo numero: '))
if resp=='Sumar':
  print('{} + {} = {}'.format(num1,num2,num1+num2))
elif resp=='Restar':
  print('{} - {} = {}'.format(num1,num2,num1-num2))
elif resp=='Multiplicar':
  print('{} * {} = {}'.format(num1,num2,num1*num2))
elif resp=='Dividir':
  print('{} / {} = {}'.format(num1,num2,num1/num2))
else:
  print('Comando no valido')
```
